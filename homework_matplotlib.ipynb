```python
#labraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# X Values
x_val= np.linspace(-7, 7, 100)
```

```python
#Sigma Values
sig_1= 1
sig_2= 0.75
sig_3= 0.5
```

```python
# Gaussian function
exp= np.exp(-0.5*(x_val/sig_1)**2)
div= sig_1*(2*np.pi)**(1/2)
sigma1= exp/div

exp_2= np.exp(-0.5*(x_val/sig_2)**2)
div_2= sig_2*(2*np.pi)**(1/2)
sigma2= exp_2/div_2

exp_3= np.exp(-0.5*(x_val/sig_3)**2)
div_3= sig_3*(2*np.pi)**(1/2)
sigma3= exp_3/div_3
```

```python
# Graphic

plt.plot(x_val, sigma1, linestyle= "-", label= "$\sigma=1.00$", color= "darkblue")
plt.plot(x_val, sigma2, linestyle= "-.", label= "$\sigma=0.75$", color= "green")
plt.plot(x_val, sigma3, linestyle= "--", label= "$\sigma=0.5$", color= "orange")

#adding axes and title
plt.title("Gaussian function for different Sigma ($\sigma$) values")
plt.xlabel("X values")
plt.ylabel("Gausian function value")

#Legend
plt.legend()
```

```python
from pathlib import Path
```

```python
# 1. Read dataset
temp_cp = Path("../CoCoLessons/data/temperature-central-park.dat")
data_temp = np.loadtxt(temp_cp)
```

```python
# 2. from Fahrenheit to Celsius

days= data_temp[:,0]
temperature = data_temp[:,1]

celsius= (5/9)*(temperature-32)
```

```python
# 3 graph

plt.plot(days, celsius, linestyle='dashed', color='blue', 
         marker= 'o', markerfacecolor='green',  label='Temperature')
plt.legend()
plt.ylabel('Temperature (ºC)')
plt.xlabel('Days')
plt.title('Temperatures Central Park January 2021')
plt.grid()
```
